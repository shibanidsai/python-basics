```python
# Start with an empty dictionary and allow a user to input marks for three subjects.
marks = {}
for i in range(3):
    subject_name = input(f"Enter subject {i+1} name: ")
    subject_mark = float(input(f"Enter marks for {subject_name}: "))
    marks[subject_name] = subject_mark
print(marks)
```

```python
# Store 9 and 9.0 as separate values in a set.
```

the output {9} from cell e181a41d is correct! Even though you added both 9 (an integer) and 9.0 (a float) to the set, Python treats them as equal values. Since sets only store unique elements, only one of them is kept in the set


```python
my_set = set()
my_set.add(9)
my_set.add(9.0)
print(my_set)
```
