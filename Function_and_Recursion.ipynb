When a recursive function lacks a base case, it calls itself indefinitely, leading to infinite recursion. In Python, this eventually raises a `RecursionError`.

```python
def fact(n):
  if( n==0 or n==1):
    return 1
  else:
    return n*fact(n-1)
```

```python
fact(5)
```

```python
#practise problems
# natural number summation
def recursive_sum(n):
  if n==0:
    return 0
  else:
    return n+recursive_sum(n-1)
```

```python
recursive_sum(5)
```

```python
#Write a recursive function to print all elements in a list.
#thinking of flow and definately abt base case-when the recursion should stop
def print_list(list,idx=0):#by putting idx=0 we make index passing while functional call as optional
  if(idx==len(list)):
    return
  print(list[idx])
  print_list(list,idx+1)
```

```python
print_list(fruits)
```
