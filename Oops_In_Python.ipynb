### Static Methods

Static methods in Python classes are methods that don't operate on the instance or the class itself. They are more like regular functions but are logically grouped within a class.

**Key Characteristics:**
* They do not receive an implicit first argument (like `self` for instance methods or `cls` for class methods).
* They can be called directly on the class (e.g., `MyClass.static_method()`) or on an instance (e.g., `my_object.static_method()`), but they don't access or modify instance or class state.
* They are defined using the `@staticmethod` decorator.

```python
class MathOperations:
    def __init__(self, value):
        self.value = value

    @staticmethod
    def add(x, y):
        return x + y

    @staticmethod
    def multiply(x, y):
        return x * y

# Calling static methods directly on the class
result_add = MathOperations.add(10, 5)
print(f"Addition result: {result_add}")

result_multiply = MathOperations.multiply(10, 5)
print(f"Multiplication result: {result_multiply}")

# Creating an instance (though not necessary to call static methods)
calc = MathOperations(20)
print(f"Value from instance: {calc.value}")

# Calling static methods through an instance (still behaves the same)
result_add_instance = calc.add(7, 3)
print(f"Addition result via instance: {result_add_instance}")
```

### Abstraction

Abstraction in Object-Oriented Programming (OOP) is the concept of hiding the complex implementation details and showing only the essential features of the object. It focuses on 'what' an object does rather than 'how' it does it. This helps in managing complexity by presenting a simpler view to the user.

**Key Characteristics:**
* **Simplification:** Reduces complexity by hiding unnecessary details.
* **Encapsulation:** Often works hand-in-hand with encapsulation (bundling data and methods that operate on the data within a single unit).
* **Focus on Interface:** Emphasizes the public interface of an object, allowing users to interact with it without needing to know its internal workings.

Let's refine your `Car` class to better demonstrate abstraction. We'll add methods like `start()`, `accelerate()`, and `brake()` that perform actions without exposing the internal state changes directly.

```python
class Car:
  def __init__(self):
    self.acc=False
    self.brk=False
    self.clutch=False
    print("Car engine created.")

  def start(self):
    self.clutch = True
    print("Car started! clutch engaged.")

  def accelerate(self):
    if self.acc:
      print("Car is accelerating...")
    else:
      print("Car is off, please start the car first.")

  def brake(self):
    self.brk = True
    print("Brakes applied. Car slowing down.")

  def stop(self):
    self.acc = False
    self.brk = False
    self.clutch = False
    print("Car stopped. All controls disengaged.")

# Demonstrating abstraction
my_car = Car()
my_car.start()
my_car.accelerate()
my_car.brake()
my_car.stop()

# The user doesn't need to know about self.acc, self.brk, self.clutch directly to use the car.
```

```python
#Practise Question 2
#Create account class with 2 atrributes balance and account no,
# create methods for debit,credit and printing balance

class Account1:
  def __init__(self,balance,acc):

      self.balance=balance
      self.accountno=acc

  def accountdetails(self):
     print(f"For account number {self.accountno}, current balance is {self.balance}")



  def debit(self,amount):
    self.balance -= amount
    print("Rs", amount, "was debited")
    print("total balance=",self.get_balance())

  def credit(self,amount):
    self.balance += amount
    print("Rs", amount, "was debited")
    print("total balance=",self.get_balance())


  def get_balance(self):
    return self.balance


acc1=Account1(10000,12345)
acc1.accountdetails()
acc1.debit(500)
acc1.credit(1500)
```

Calling methods like acc1.debit() demonstrates abstraction because it allows you to interact with the Account1 object without needing to know the complex internal details of how the debit operation is performed. When you call acc1.debit(500), you are simply stating what you want to do (debit 500 units from the account). You don't need to understand how the balance attribute is updated, or any other internal logic within the debit method. This hiding of implementation details and exposure of only essential functionality is the core principle of abstraction.

```python
# single level inheritance

class Car:
  color="blue"
  @staticmethod #make it static as we are not passing variables
  def start():
    print("Car has started")

  @staticmethod
  def stop():
    print("Car has stopped")

class ToyotaCar(Car):
  def __init__(self,name): # Corrected: __init__ with two underscores
    self.name=name

car1= ToyotaCar("fortuner")
car2= ToyotaCar("prius")

print(car1.name)
print(car1.start())
```

```python
# Multilevel inheritance
class Car:
  color="blue"
  @staticmethod #make it static as we are not passing variables
  def start():
    print("Car has started")

  @staticmethod
  def stop():
    print("Car has stopped")

class ToyotaCar(Car):
  def __init__(self,name): # Corrected: __init__ with two underscores
    self.name=name

class Fortuner(ToyotaCar):
  def __init__(self,type):
    self.type=type

car1= Fortuner("diesel")

print(car1.type)
print(car1.start())
```

```python
#usage of property
class Student:
  def __init__(self,phy,chem,math):
    self.phy=phy
    self.chem=chem
    self.math=math

  @property
  def percentage(self):
    return str((self.phy+self.chem+self.math)/3)+"%"


stu1= Student(98,97,99)
print(stu1.percentage)
stu1.phy=82
print("%changed:",stu1.percentage)
```

That's a great question about how object-oriented programming works!

Inside a method of a class, self is a special parameter that refers to the instance of the class on which the method was called. So, when you write num1.add(num2):

num1 becomes self within the add method.
num2 is the argument passed to the add method.
Therefore, self.real refers to the real attribute of the num1 object (the object on which the add method was invoked), and num2.real refers to the real attribute of the num2 object (the object passed as an argument). This allows the method to correctly add the real parts of the two complex numbers.
