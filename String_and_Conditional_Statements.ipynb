elif (short for 'else if') is used when you have multiple conditions to check sequentially. It allows you to specify additional conditions after an initial if statement, and only if the preceding if or elif conditions were false. It's typically used in combination with if and optionally followed by a final else for a default action if none of the preceding conditions are met.


```python
marks = int(input("input your marks to get grade "))

if marks > 90:
  print("grade A+")
elif marks >= 80 and marks <= 89:
  print("Grade A")
elif marks >= 70 and marks <= 79:
  print("Grade B")
else:
  print("Grade F")

  #what I had missed..takig input as int and there is between statement but there is > and <
  # also indentation should be good
```
