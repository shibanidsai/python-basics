Typecasting a string to a tuple using tuple() converts the string into a tuple of its individual characters. For example, tuple("abc") becomes ('a', 'b', 'c'). This works only when the input is an iterable, like a string or list. A single number (e.g., tuple(5)) cannot be converted directly because numbers are not iterable.same happens for casting by set,list,dict
