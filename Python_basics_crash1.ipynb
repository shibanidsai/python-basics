we will learn how to use map instead of for looop which is much effiencent way to return new list of squared numbers


```python
intlist=[1,2,3,4,5,6,7]
```

```python
# creating a new list to store square of numbers
squaredList=[]
for i in intlist:
  squaredList.append(pow(i,2))

print(squaredList)

```

```python
def square_me(i):
  return i*i
```

```python
squaredList2=map(square_me,intlist)
list(squaredList2) #print(squarelist) does not work
```

## using lamda function for Map and filter


```python
intlist1=[5,6,7,8,9]
squaredlistlamda=map(lambda x:x*x,intlist1 )
list(squaredlistlamda)
```

```python
even=filter(lambda x:x%2==0,intlist1)
list(even)
```

above lambda function is more use but good to know below iterations as well

In Python, a function can return multiple values using a tuple. When values are separated by commas, Python automatically packs them into a tuple.



```python
# returning 2 elements in function using tuple
from statistics import mean, median

def getMeanAndMedian( listNum ):
    return mean(listNum), median(listNum)
```

```python
# The returned tuple (value1, value2) is unpacked.

# mean variable gets the first value.

# median variable gets the second value.
#mean, median = getMeanAndMedian(randomList)
```

Python does tuple packing and unpacking automatically.

Packing: return a, b → (a, b)

Unpacking: x, y = (a, b)

```python
mean,median=getMeanAndMedian( randomList )
```

```python
print( "Mean: ", mean, " Median: ", median)
```
