```python
heroes = ['ironman', 'thor', 'superman', 'batman']
target = 'superman'

```

```python
# break: Stopping a search once an item is found
for hero in heroes:
  if hero==target:
    print("Hero found!Stop search immediately")
    break
  print(f"Checking {hero}")
```

```python
# continue: Printing only odd numbers between 1 and 10

i=0
while i<10:
  i+=1
  if i%2 ==0:
    continue
  print(i)
```

```python
# for with else loop
for el in heroes:
  print(el)
else:
  print("All heroes processed successfully")
```

```python
#pass statement
def my_function():
  pass # This function does nothing yet, but avoids a syntax error

if True:
  pass # This conditional does nothing, but avoids a syntax error

for i in range(3):
  pass # This loop does nothing for each iteration

print("Pass statement executed without error.")
```

The `pass` statement in Python is a null operation. When it is executed, nothing happens. It is useful as a placeholder when a statement is syntactically required but you do not want any code to execute. This is common in situations where you are defining a function, class, or conditional block that you plan to implement later, or if you need an empty block for some logical reason.

```python
#sum of first n numbers using while
n=int(input("How many numbers sum you want? "))
total_sum=0
i=1
while i<=n:
    total_sum +=i
    i+=1
print(f"Total sum:",total_sum)
```

```python
#Factorial of a number
n=int(input("which numbers factorial you want?"))
fact=1
for i in range(1,n+1):
  fact*=i
print(f"factorial of {n}:",fact)
```

```python
#DS Usecase

# Simulating processing a large dataset (e.g., 100,000 data entries)
num_entries = 100000

print(f"Processing {num_entries} data entries...")

# Using a loop to process each entry without writing repetitive code
for i in range(1, num_entries + 1):
  # In a real scenario, you'd perform an operation here,
  # such as reading data, performing a calculation, or sending an email.
  if i % 10000 == 0: # Print a progress update every 10,000 entries
    print(f"  Processed {i} entries...")

print("Finished processing all entries.")
```

This code snippet demonstrates how a simple `for` loop can iterate over `100,000` simulated data entries. Instead of writing `100,000` lines of code for each entry, the loop allows you to define the processing logic once, and it gets applied to every item. This makes your code scalable, efficient, and adheres to the 'Don't Repeat Yourself' (DRY) principle, which is essential for large-scale data processing or automation tasks like sending emails.

```python
#Search for a number x in this tuple using loop:
tuple1=(1, 4, 9, 16, 25, 36, 49, 64, 81,100)
x=36
idx=0
found_idx = -1 # Use -1 as a sentinel value to indicate if x was found
for i in tuple1:
  if i == x: # Correct condition: check if current element 'i' equals target 'x'
    found_idx = idx # Store the index where x was found
    break # Stop search immediately once found
  idx += 1 # Increment index for each element checked

if found_idx != -1: # Check if x was found
  print(f"number {x} is found at {found_idx}")
else:
  print(f"number {x} not found in the tuple.")
```

```python
# Search for a number x in this tuple using a while loop:
tuple1=(1, 4, 9, 16, 25, 36, 49, 64, 81,100)
x=49
idx=0
found_idx = -1 # Use -1 as a sentinel value

while idx < len(tuple1):
  if tuple1[idx] == x:
    found_idx = idx
    break
  idx += 1

if found_idx != -1:
  print(f"Number {x} is found at index {found_idx}")
else:
  print(f"Number {x} not found in the tuple.")
```
